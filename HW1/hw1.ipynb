<h2>Q2 - Low pass filters & high pass filters</h2>

```python
img = cv2.imread("input2.png",0);
```

```python
img.shape
```

```python
f = numpy.fft.fft2(img)
fshift = numpy.fft.fftshift(f)
magnitude_spectrum = numpy.log(numpy.abs(fshift))
magnitude_spectrum = numpy.array(magnitude_spectrum * 15, dtype = numpy.uint8)
```

```python
cv2.imshow('image',magnitude_spectrum)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
rows, cols = magnitude_spectrum.shape
crow,ccol = rows/2 , cols/2
```

```python
fshift[crow-10:crow+10, ccol-10:ccol+10] = 0   # what is fshift doing?
f_ishift = numpy.fft.ifftshift(fshift)
img_back = numpy.fft.ifft2(f_ishift)
img_back = numpy.abs(img_back)
```

```python
fshift[110,110]
```

```python
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])
plt.show()
```

<h2>LPF</h2>

```python
rows, cols = magnitude_spectrum.shape
crow,ccol = rows/2 , cols/2
fshift[0:crow-10, :] = 0
fshift[crow+10:, :] = 0
fshift[:, 0:ccol-10] = 0
fshift[:, ccol+10:] = 0
f_ishift = numpy.fft.ifftshift(fshift)
img_back = numpy.fft.ifft2(f_ishift)
img_back = numpy.abs(img_back)
```

```python
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])
plt.show()
```

<h2>Deconvolution</h2>

```python
img = cv2.imread("blurred2.exr", cv2.IMREAD_ANYCOLOR | cv2.IMREAD_ANYDEPTH)
```

```python
img.shape
```

```python
gk = cv2.getGaussianKernel(21,5)
gk = gk * gk.T

def ft(im, newsize=None):
    dft = np.fft.fft2(np.float32(im),newsize)
    return np.fft.fftshift(dft)

def ift(shift):
    f_ishift = np.fft.ifftshift(shift)
    img_back = np.fft.ifft2(f_ishift)
    return np.abs(img_back)
```

```python
imf = ft(img, (img.shape[0],img.shape[0])) # make sure sizes match
gki = numpy.linalg.inv(gk)
gkfi = ft(gki, (img.shape[0],img.shape[0])) # so we can multiple easily
```

```python
#gkfi = numpy.linalg.inv(gkf)
```

```python
imconvf =  imf * gkfi
blurred = np.log(ift(imconvf))
```
